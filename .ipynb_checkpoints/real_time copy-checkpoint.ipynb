```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import IPython.display as ipd  # To play sound in the notebook
import os
from tqdm import tqdm
import datetime
import tensorflow as tf
import pyaudio
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from matplotlib import animation as animation
from keras_tuner.tuners import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
from keras.models import load_model
from pydub import AudioSegment, effects
from sklearn.utils import shuffle
import noisereduce as nr
import io
import wave
```

```python
EMOTIONS = {
    0: 'neutral',
    1: 'calm',
    2: 'happy',
    3: 'sad',
    4: 'angry',
    5: 'fear',
    6: 'disgust',
    7: 'surprise'
}
emo_list = list(EMOTIONS.values())
max_length = 228864
frame_length = 2048
hop_length = 512
CHUNK = 512
sr = 48000
RECORD_SECONDS = 4.8
FORMAT = pyaudio.paInt32
CHANNELS = 1
WAVE_OUTPUT_FILE = "temp.wav"
```

```python
def get_features(data):

    scaler = StandardScaler()

    # Initialize data lists
    rms = []
    zcr = []
    mfcc = []
    sc = []
    cf = []
    mel = []

    # Fetch the sample rate.
    # rawsound, sr = librosa.load(path=path, duration=3, offset=0.2)
    # sr (the sample rate) is used for librosa's MFCCs. '_' is irrelevant.
    # Load the audio file.
    rawsound = AudioSegment.from_file(data)
    # rawsound = AudioSegment(data)
    # Normalize the audio to +5.0 dBFS.
    normalizedsound = effects.normalize(rawsound, headroom=5.0)
    # Transform the normalized audio to np.array of samples.
    normal_audio = np.array(normalizedsound.get_array_of_samples(),
                            dtype='float32')
    # Trim silence from the beginning and the end.
    trimmed_audio, index = librosa.effects.trim(y=normal_audio, top_db=30)
    # print(len(trimmed_audio))
    # Pad for duration equalization.
    print(len(trimmed_audio))
    padded_audio = np.pad(trimmed_audio, (0, max_length - len(trimmed_audio)),
                          'constant')
    # Noise reduction.
    final_audio = nr.reduce_noise(y=padded_audio, sr=sr)  #updated 03/03/22

    # Features extraction
    f1 = librosa.feature.rms(
        y=final_audio, frame_length=frame_length,
        hop_length=hop_length).T  # Energy - Root Mean Square
    f2 = librosa.feature.zero_crossing_rate(y=final_audio,
                                            frame_length=frame_length,
                                            hop_length=hop_length,
                                            center=True).T  # ZCR
    f3 = librosa.feature.mfcc(y=final_audio,
                              sr=sr,
                              n_mfcc=20,
                              hop_length=hop_length).T  # MFCC

    f4 = librosa.feature.spectral_centroid(
        y=final_audio, sr=sr, hop_length=hop_length).T  # pectral centroid

    f5 = librosa.feature.chroma_stft(y=final_audio, sr=sr).T  # chroma feature
    f6 = librosa.feature.melspectrogram(y=final_audio,
                                        n_fft=frame_length,
                                        hop_length=hop_length).T

    # Filling the data lists

    rms.append(scaler.fit_transform(f1))
    zcr.append(scaler.fit_transform(f2))
    mfcc.append(scaler.fit_transform(f3))
    sc.append(scaler.fit_transform(f4))
    cf.append(scaler.fit_transform(f5))
    mel.append(scaler.fit_transform(f6))
    f_rms = np.asarray(rms)
    f_zcr = np.asarray(zcr)
    f_mfccs = np.asarray(mfcc)
    f_sc = np.asarray(sc)
    f_cf = np.asarray(cf)
    f_mel = np.asarray(mel)
    # Concatenating all features to 'X' variable.
    features = np.concatenate((f_zcr, f_rms, f_mfccs, f_sc, f_cf, f_mel),
                              axis=2)
    return features
```

```python
model = load_model('Models/LSTM_Best_Model.hdf5')
```

```python
def is_silent(data):
    # Returns 'True' if below the 'silent' threshold
    return max(data) < 200
```

```python
def animate(frame):
    global emo_list, pred_np
    plt.cla()
    plt.bar(emo_list, pred_np, color='darkturquoise')
    plt.tight_layout()
```

```python
import pyaudio
import wave
from array import array
import struct
import time

RATE = sr

WAVE_OUTPUT_FILE = "temp.wav"

# Open an input channel
p = pyaudio.PyAudio()
stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# fig = plt.figure()

colors = ['red', 'yellow', 'blue', 'green', 'black']

# Initialize a non-silent signals array to state "True" in the first 'while' iteration.
# data = array('h', np.random.randint(size=512, low=0, high=500))
# ani = animation.FuncAnimation(plt.gcf(), animate)

while True:
    data = array("l",stream.read(CHUNK, exception_on_overflow=False))
    if is_silent(data) == False:

        frames = []

        timesteps = int(RATE / CHUNK * RECORD_SECONDS)  # => 339

        frames.append(data)

        wf = wave.open(WAVE_OUTPUT_FILE, 'wb')
        wf.setnchannels(CHANNELS)
        wf.setsampwidth(p.get_sample_size(FORMAT))
        wf.setframerate(RATE)
        wf.writeframes(b''.join(frames[:timesteps]))

        del frames[:timesteps]
        try:

            x = get_features(
                WAVE_OUTPUT_FILE)  # 'output.wav' file preprocessing.
        except Exception as e:
            print(e)
            continue
        # Model's prediction => an 8 emotion probabilities array.
        predictions = model.predict(x, use_multiprocessing=True)
        pred_list = list(predictions)
        pred_np = np.squeeze(np.array(pred_list).tolist(),
                             axis=0)  # Get rid of 'array' & 'dtype' statments.

        max_emo = np.argmax(predictions)

        print('max emotion:', EMOTIONS.get(max_emo, -1))
#         plt.cla()
#         plt.bar(emo_list, pred_np, color='darkturquoise')
#         plt.tight_layout()
#         plt.ylabel("Probabilty (%)")

#         plt.show()

# toc = time.perf_counter()
stream.stop_stream()
stream.close()
p.terminate()
wf.close()
print('** session ended')
```
